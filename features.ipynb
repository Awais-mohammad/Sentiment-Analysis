### Count of words with +ve,-ve and ~ sentiments in docs

```python
# Pos tagging
tokened_Sen=[]
def POS_taggs():
    for i in range(0, len(finale)):
        token = word_tokenize(finale[i])
        afterTag = nltk.pos_tag(token)
        tokened_Sen.append(afterTag)
    return tokened_Sen
```

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
POS_taggs()
```

```python
positiveWords=[]
negativeWords=[]
neutralWords=[]
for i in range(0,len(tokened_Sen)):
    posCount=0
    negCount=0
    neuCount=0
    for j in range(0,len(tokened_Sen[i])):
        synsets=wn.synsets(tokened_Sen[i][j][0],pos=get_wordnet_pos(tokened_Sen[i][j][1]))
        if synsets:
            synset = synsets[0]
            swn_synset = swn.senti_synset(synset.name())
            if swn_synset.pos_score() > swn_synset.neg_score() and swn_synset.pos_score() > swn_synset.obj_score():
                posCount+=1
            elif swn_synset.neg_score() > swn_synset.pos_score() and swn_synset.neg_score() > swn_synset.obj_score():
                negCount+=1
            elif swn_synset.obj_score() > swn_synset.pos_score() and swn_synset.obj_score() > swn_synset.neg_score():
                neuCount+=1
    positiveWords.append(posCount)
    negativeWords.append(negCount)
    neutralWords.append(neuCount)
```

```python
newDF=pd.DataFrame( {'PositiveWords': positiveWords,
     'NegativeWords': negativeWords,
     'NeutralWords': neutralWords
    })
```

```python
newDF.insert(len(newDF.columns),'Sentiments',df['Sentiments'])
```

```python
newDF.to_csv('.\\Features\countofWords.csv',index=False)
```

# Implementation of bag of words to get top 10 reoccuring words

```python
a=pd.Series(finale)
a=a.map(''.join)
```

```python
reviews=[]
for i in range(0,len(a)):
    tokens=word_tokenize(a[i])
    sentence=[]
    joinSentence=[]
    for word in tokens:
        if word[0] != word[1]:
            if  word[1] != word[2]:
                sentence.append(word)
            joinSentence=" ".join(sentence)
    reviews.append(joinSentence)
```

```python
covid=pd.Series(reviews)
omicronReviews=covid.map(''.join)
```

```python
allReviews=[]
for i in range(0,len(omicronReviews)):
    token=word_tokenize(omicronReviews[i])
    words=[]
    sentence=[]
    for word in token:
        if len(word) > 3:
            words.append(word)
        sentence=" ".join(words)
    allReviews.append(sentence)
```

```python
covid=pd.Series(allReviews)
threeMore_Reviews=covid.map(''.join)
```

```python
CountVec = CountVectorizer(ngram_range=(1,1),stop_words='english',min_df=10)
```

```python
Count_data = CountVec.fit_transform(threeMore_Reviews)
```

```python
cv_dataframe=pd.DataFrame(Count_data.toarray(),columns=CountVec.get_feature_names())
```

```python
add={}
for i in cv_dataframe.columns:
    add[i]=cv_dataframe[i].sum()
```

### Word2vec implementatoion

```python
!pip install gensim
```

```python
tokenedData=rTxt.apply(lambda x:x.split())
```

```python
model = Word2Vec(sentences=tokenedData, vector_size=350, window=5,workers=4,sg=1)
```

```python
model.train(tokenedData,total_examples=len(tokenedData),epochs=14)
```

```python
def word_vector(tokens, size):   
    vec = np.zeros(size).reshape((1, size))
    count = 0
    for word in tokens:
        try:
            vec += model.wv[word].reshape((1, size))
            count += 1.
        except KeyError:  # handling the case where the token is not in vocabulary
            continue
    if count != 0:
        vec /= count
    return vec
```

```python
wordvec_arrays = np.zeros((len(rTxt), 300))                                              #Preparing word2vec feature set
for i in range(len(rTxt)):
    wordvec_arrays[i,:] = word_vector(rTxt[i], 300)
wordvec_df = pd.DataFrame(wordvec_arrays)
wordvec_df.shape
```

```python
wordvec_df.insert(len(df.columns),'Sentiments',df['Sentiments'])
```
